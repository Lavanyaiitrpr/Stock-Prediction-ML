Momentum( 10 days) = Cn - C n-10

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

data = yf.download('TATAMOTORS.NS', start='2023-01-01', end='2023-02-01')
df = pd.DataFrame(data['Close'], index=data.index)

mtm = [0] * 10  # Initialize first 10 values with 0

for i in range(10, len(df)):
    mtm_val = df.iloc[i][0] - df.iloc[i - 10][0]  # Calculate the difference between current price and price from 10 periods ago
    mtm.append(mtm_val)

mtm_ser = pd.Series(mtm, index=df.index, name='Momentum')

df['Momentum'] = mtm_ser
print(df)

#Plotting the garph
plt.plot(df.index, df['Momentum'],label='Momentum', color='blue')
plt.grid()

# Adding the title
plt.title("TATA MOTORS Momentum")

# Adding the labels
plt.ylabel("Momentum")
plt.xlabel("Date")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.legend()
plt.show()
```

# Moving Average

SIMPLE MOVING AVERAGE

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

#Fetching data from yfinance library
df = yf.download('TATAMOTORS.NS',start='2024-01-01')
#Date in YYYY-MM-DD format
mavg=[0]*2
for i in range(3,len(df)+1,1):
    data = pd.DataFrame(df.Close[i-10:i],)
    #taking period of moving average of 10 days
    mavg.append(data.mean())

df['Mavg'] = pd.Series(mavg,index= df.index)


plt.plot(df.index,df['Mavg'],label='Moving Average', color = 'blue')
plt.grid()

# Adding the title
plt.title("TATA MOTORS")

# Adding the labels
plt.ylabel("Moving Average")
plt.xlabel("TimeSeries")
plt.xticks(rotation=90)
plt.legend()
plt.show()

```
